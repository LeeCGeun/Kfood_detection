```python
import pyautogui
import time

# ë¼ë²¨ ë²”ìœ„ ì„¤ì • (005 ~ 149)
start_label = 1
end_label = 149

# ì¢Œí‘œ ì„¤ì • (ìˆ˜ì •ëœ ì¢Œí‘œ)
first_position = (1108, 318)  # ì²« ë²ˆì§¸ ì»¤ì„œ ì´ë™ ìœ„ì¹˜
click_position1 = (1123, 363)  # ì²« ë²ˆì§¸ í´ë¦­

click_position2 = (516, 89)  # ë‘ ë²ˆì§¸ í´ë¦­
input_position = (385, 555)  # ìˆ«ì ì…ë ¥í•  ìœ„ì¹˜
select_all_position = (304, 235)  # Ctrl + A ëˆ„ë¥¼ ìœ„ì¹˜
final_position = (751, 589)  # ë§ˆì§€ë§‰ í´ë¦­ ìœ„ì¹˜

# ë”œë ˆì´ ì„¤ì •
delay = 3  # ì²« ì´ë™ í›„ ëŒ€ê¸° ì‹œê°„ (ì´ˆ)
input_delay = 2  # ê° ë™ì‘ ê°„ ë”œë ˆì´
wait_time = 420 # 15ë¶„(900ì´ˆ) ëŒ€ê¸° ì‹œê°„

for i in range(start_label, end_label + 1):
    label = f"{i}"  # 005 ~ 149 ìˆ«ì í¬ë§· ë§ì¶”ê¸°

    print(f"ğŸ“Œ {label} ì²˜ë¦¬ ì‹œì‘...")
    time.sleep(delay)
    # ì²« ë²ˆì§¸ ì»¤ì„œ ì´ë™ í›„ ëŒ€ê¸°
    pyautogui.moveTo(first_position)
    time.sleep(delay)

    # ì²« ë²ˆì§¸ í´ë¦­
    pyautogui.click(click_position1)
    time.sleep(input_delay)

    # ë‘ ë²ˆì§¸ í´ë¦­
    pyautogui.click(click_position2)
    time.sleep(input_delay)

    # ìˆ«ì ì…ë ¥
    pyautogui.click(input_position)
    time.sleep(input_delay)
    pyautogui.write(label)  # 005 ~ 149 ì…ë ¥
    pyautogui.press('enter')  # Enter ì…ë ¥
    time.sleep(input_delay)

    # Ctrl + A ì…ë ¥
    pyautogui.click(select_all_position)
    time.sleep(input_delay)
    pyautogui.hotkey('ctrl', 'a')  # Ctrl + A
    time.sleep(input_delay)

    # ë§ˆì§€ë§‰ ìœ„ì¹˜ í´ë¦­
    pyautogui.click(final_position)
    time.sleep(input_delay)

    print(f"âœ… {label} ì²˜ë¦¬ ì™„ë£Œ! 15ë¶„ ëŒ€ê¸° ì¤‘...")
    
    if i < end_label:
        time.sleep(wait_time)  # 15ë¶„ ëŒ€ê¸°

print("ğŸ‰ ëª¨ë“  ë§¤í¬ë¡œ ì‘ì—… ì™„ë£Œ!")
```

```python
import json
import re
from pathlib import Path

# íŒŒì¼ ê²½ë¡œ
file_path = Path("magnesium_50.json")

# parse_value í•¨ìˆ˜ ë‚´ì— ë¬¸ìì—´ í™•ì¸ ì¡°ê±´ ì¶”ê°€í•˜ì—¬ float ì…ë ¥ ì²˜ë¦¬
def parse_value(value):
    if isinstance(value, (int, float)):
        return float(value)  # ì´ë¯¸ ìˆ«ìë©´ ê·¸ëŒ€ë¡œ ë°˜í™˜

    value = value.strip().replace(",", ".")
    match = re.match(r"([\d.]+)\s*([^\d\s]*)", value)
    if match:
        num_str = match.group(1).rstrip(".")
        try:
            number = float(num_str)
        except ValueError:
            return value

        unit = match.group(2)

        if unit.startswith(("Âµg", "Î¼g")):
            return number * 0.001
        elif unit == "mg":
            return number
        elif unit == "g":
            return number * 1000
        elif unit == "kcal":
            return number
    return value

# í‚¤ ì´ë¦„ ì •ì œ
key_rename = {
    "ì—´ëŸ‰": "ì—ë„ˆì§€",
    "ì…€ë Œ": "ì…€ë ˆëŠ„"
}

# JSON íŒŒì¼ ë¡œë”©
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# ì „ì²˜ë¦¬ ìˆ˜í–‰
for product in data:
    if "nutrition" in product:
        nutrition = product["nutrition"]
        new_nutrition = {}
        for k, v in nutrition.items():
            key = key_rename.get(k.strip(), k.strip())
            val = parse_value(v)
            new_nutrition[key] = val
        product["nutrition"] = new_nutrition

# ì €ì¥
output_path = "magnesium_unit_converted_float.json"
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
import json

# ê¸°ì¤€ ì˜ì–‘ì†Œ ë¦¬ìŠ¤íŠ¸ (ë‹¨ìœ„ ì œê±°)
target_nutrients = [
    "ì—ë„ˆì§€", "íƒ„ìˆ˜í™”ë¬¼", "ì‹ì´ì„¬ìœ ", "ë‹¨ë°±ì§ˆ", "ë¦¬ë†€ë ˆì‚°", "ì•ŒíŒŒ-ë¦¬ë†€ë Œì‚°",
    "EPA+DHA", "ë©”í‹°ì˜¤ë‹Œ", "ë¥˜ì‹ ", "ì´ì†Œë¥˜ì‹ ", "ë°œë¦°", "ë¼ì´ì‹ ",
    "í˜ë‹ì•Œë¼ë‹Œ+í‹°ë¡œì‹ ", "íŠ¸ë ˆì˜¤ë‹Œ", "íŠ¸ë¦½í† íŒ", "íˆìŠ¤í‹°ë”˜", "ë¹„íƒ€ë¯¼A",
    "ë¹„íƒ€ë¯¼D", "ë¹„íƒ€ë¯¼E", "ë¹„íƒ€ë¯¼K", "ë¹„íƒ€ë¯¼C", "ë¹„íƒ€ë¯¼B1",
    "ë¹„íƒ€ë¯¼B2", "ë‚˜ì´ì•„ì‹ ", "ë¹„íƒ€ë¯¼B6", "ë¹„íƒ€ë¯¼B12", "ì—½ì‚°",
    "íŒí† í…ì‚°", "ë¹„ì˜¤í‹´", "ì¹¼ìŠ˜", "ì¸", "ë‚˜íŠ¸ë¥¨", "ì—¼ì†Œ",
    "ì¹¼ë¥¨", "ë§ˆê·¸ë„¤ìŠ˜", "ì² ", "ì•„ì—°", "êµ¬ë¦¬", "ë§ê°„",
    "ìš”ì˜¤ë“œ", "ì…€ë ˆëŠ„", "ëª°ë¦¬ë¸Œë´", "í¬ë¡¬"
]

# ì˜ì–‘ì†Œ ì´ë¦„ì—ì„œ ë‹¨ìœ„ ì œê±°
def strip_unit(nutrient_name):
    return nutrient_name.split('(')[0].split('[')[0].strip()

# íŒŒì¼ ë¡œë”©
with open('magnesium_unit_converted_float.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# nutrition í•„í„°ë§ + ì‚­ì œëœ í•­ëª© ì¶”ì 
for item in data:
    nutrition = item.get("nutrition", {})
    filtered_nutrition = {}
    removed_keys = []

    for k, v in nutrition.items():
        if strip_unit(k) in target_nutrients:
            filtered_nutrition[k] = v
        else:
            removed_keys.append(k)

    item["nutrition"] = filtered_nutrition

    if removed_keys:
        print(f"[{item['name']}] ì‚­ì œëœ ì˜ì–‘ì†Œ: {removed_keys}")

# ì €ì¥
with open('magnesium_preprocessed.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
