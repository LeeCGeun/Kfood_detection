```python
import pyautogui
import time

# 라벨 범위 설정 (005 ~ 149)
start_label = 1
end_label = 149

# 좌표 설정 (수정된 좌표)
first_position = (1108, 318)  # 첫 번째 커서 이동 위치
click_position1 = (1123, 363)  # 첫 번째 클릭

click_position2 = (516, 89)  # 두 번째 클릭
input_position = (385, 555)  # 숫자 입력할 위치
select_all_position = (304, 235)  # Ctrl + A 누를 위치
final_position = (751, 589)  # 마지막 클릭 위치

# 딜레이 설정
delay = 3  # 첫 이동 후 대기 시간 (초)
input_delay = 2  # 각 동작 간 딜레이
wait_time = 420 # 15분(900초) 대기 시간

for i in range(start_label, end_label + 1):
    label = f"{i}"  # 005 ~ 149 숫자 포맷 맞추기

    print(f"📌 {label} 처리 시작...")
    time.sleep(delay)
    # 첫 번째 커서 이동 후 대기
    pyautogui.moveTo(first_position)
    time.sleep(delay)

    # 첫 번째 클릭
    pyautogui.click(click_position1)
    time.sleep(input_delay)

    # 두 번째 클릭
    pyautogui.click(click_position2)
    time.sleep(input_delay)

    # 숫자 입력
    pyautogui.click(input_position)
    time.sleep(input_delay)
    pyautogui.write(label)  # 005 ~ 149 입력
    pyautogui.press('enter')  # Enter 입력
    time.sleep(input_delay)

    # Ctrl + A 입력
    pyautogui.click(select_all_position)
    time.sleep(input_delay)
    pyautogui.hotkey('ctrl', 'a')  # Ctrl + A
    time.sleep(input_delay)

    # 마지막 위치 클릭
    pyautogui.click(final_position)
    time.sleep(input_delay)

    print(f"✅ {label} 처리 완료! 15분 대기 중...")
    
    if i < end_label:
        time.sleep(wait_time)  # 15분 대기

print("🎉 모든 매크로 작업 완료!")
```

### 영양성분 어떤 것 있는지

```python
import json
from collections import Counter

with open('magnesium_50.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 영양 성분 키 카운트
nutrition_counter = Counter()

for item in data:
    nutrition = item.get("nutrition", {})
    for nutrient in nutrition.keys():
        nutrition_counter[nutrient] += 1

print("철분 영양소 종류:\n")
for nutrient, count in nutrition_counter.most_common():
    print(f"{nutrient}: {count}개")
```

```python
import json
import re
from pathlib import Path

# 파일 경로
file_path = Path("magnesium_50.json")

# parse_value 함수 내에 문자열 확인 조건 추가하여 float 입력 처리
def parse_value(value):
    if isinstance(value, (int, float)):
        return float(value)  # 이미 숫자면 그대로 반환

    value = value.strip().replace(",", ".")
    match = re.match(r"([\d.]+)\s*([^\d\s]*)", value)
    if match:
        num_str = match.group(1).rstrip(".")
        try:
            number = float(num_str)
        except ValueError:
            return value

        unit = match.group(2)

        if unit.startswith(("µg", "μg")):
            return number * 0.001
        elif unit == "mg":
            return number
        elif unit == "g":
            return number * 1000
        elif unit == "kcal":
            return number
    return value

# 키 이름 정제
key_rename = {
    "열량": "에너지",
    "셀렌": "셀레늄"
}

# JSON 파일 로딩
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 전처리 수행
for product in data:
    if "nutrition" in product:
        nutrition = product["nutrition"]
        new_nutrition = {}
        for k, v in nutrition.items():
            key = key_rename.get(k.strip(), k.strip())
            val = parse_value(v)
            new_nutrition[key] = val
        product["nutrition"] = new_nutrition

# 저장
output_path = "magnesium_unit_converted_float.json"
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```

### 영양소 기준 43개 제외한 나머지 영양소 삭제

```python
import json

# 기준 영양소 리스트 (단위 제거)
target_nutrients = [
    "에너지", "탄수화물", "식이섬유", "단백질", "리놀레산", "알파-리놀렌산",
    "EPA+DHA", "메티오닌", "류신", "이소류신", "발린", "라이신",
    "페닐알라닌+티로신", "트레오닌", "트립토판", "히스티딘", "비타민A",
    "비타민D", "비타민E", "비타민K", "비타민C", "비타민B1",
    "비타민B2", "나이아신", "비타민B6", "비타민B12", "엽산",
    "판토텐산", "비오틴", "칼슘", "인", "나트륨", "염소",
    "칼륨", "마그네슘", "철", "아연", "구리", "망간",
    "요오드", "셀레늄", "몰리브덴", "크롬"
]

# 영양소 이름에서 단위 제거
def strip_unit(nutrient_name):
    return nutrient_name.split('(')[0].split('[')[0].strip()

# 파일 로딩
with open('magnesium_unit_converted_float.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# nutrition 필터링 + 삭제된 항목 추적
for item in data:
    nutrition = item.get("nutrition", {})
    filtered_nutrition = {}
    removed_keys = []

    for k, v in nutrition.items():
        if strip_unit(k) in target_nutrients:
            filtered_nutrition[k] = v
        else:
            removed_keys.append(k)

    item["nutrition"] = filtered_nutrition

    if removed_keys:
        print(f"[{item['name']}] 삭제된 영양소: {removed_keys}")

# 저장
with open('magnesium_preprocessed.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
